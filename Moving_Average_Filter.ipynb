```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Filter class
class moving_filter: 
  arr = []
  size = 0

  def __init__(self,num):
    self.arr = []
    self.size = num
    assert self.size > 0 , "x should be grater than '0'"

  def new_data(self,entry):
    self.arr.append(entry)
    if(len(self.arr) > self.size):
      self.arr.pop(0)
  
  def get_average(self):
    return sum(self.arr)/len(self.arr)

  def __del__(self):
    print("Filter Destroyed")
```

```python
#pure signal of sine wave
nparray1 = np.sin(np.linspace(0,5*np.pi,num = 180*5))
plt.plot(nparray1)
plt.show()
```

```python
#generate random guassian Noise
noise = np.sin(np.random.normal(0,0.3,180*5))
plt.plot(noise)
plt.show()
```

```python
#add the noise to the pure signal
added = noise + nparray1
plt.plot(added)
plt.show()
```

```python
filter_obj = moving_filter(5)
res = []
for i in range (0,180*5):
  filter_obj.new_data(added[i])
  res.append(filter_obj.get_average())

plt.plot(res)
plt.show()
```
